```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# NFW density profile
# rho_DM = 10^4 rho_c / [(r/Rs)(1+r/Rs)^2], beta =1 

# Axion field
# a(r,t)=sqrt(2 rho_DM)/m_a * cos(m_a t)

#COMA DM PARAMETERS : 
rho_0_SI = 1.74 * 1e-2 * 6.77 * 1e-20 #kg/m3
Rs_kpc = 12.5 #kpc 
```

```python
#natural unit covnersions 
#1 kg/m3 = 4.3101 * 10^-21 GeV^4
# 1 kpc = 1.5639 * 10^41 GeV^-1

rho_0 = rho_0_SI * 4.3 *1e-21 * 1e36 #eV^4
Rs = Rs_kpc * 1.5 *1e41 *1e-9 #ev^-1
```

```python
r = np.logspace(-2,3,1000)   # 0.01 kpc to 1000 kpc
```

```python
#Mass 
m_a = 1e-31#eV

# radius range
Rmax = 400 #kpc    - 400 kpc 
r_normal = np.linspace(0, 2*Rmax, 5000)
centre = Rmax

#Density profile
def rho_DM(r): # NFW density
      x = np.abs(r_normal - centre) / Rs_kpc
      return (rho_0 / ((x) * (1 + x)**(2)))

#axion field 
def a(r):
    return np.sqrt(2 * rho_DM(r)) / m_a

# Plot
plt.figure(figsize=(7,5))
plt.loglog(r, a(r_normal))

plt.xlabel(r'$r\ \mathrm{[kpc]}$')
plt.ylabel(r'$a_0(r)$')
plt.title(r'Axion field profile from NFW density')
plt.grid(True, which='both', alpha=0.3)

plt.show()
```

```python
x = np.abs(r_normal - centre) / Rs_kpc
#derivatives of a , in static, i.e taking cos =1 , sin=1
def a_dot(r):
    return ( -1* np.sqrt(2 * rho_0) * (x)**-0.5 * (1 +x)**-1  )

def a_prime(r):
    return ( ( np.sqrt(2 * rho_0) / ( m_a * Rs) ) * ((-0.5 * (x)**-1.5 * (1 + x)**-1) + (-1* (x)**-0.5 * (1 + x)**-2)))

plt.figure(figsize=(8, 6))

plt.loglog(r,a(r_normal),lw=2,label=r'$a$')

plt.loglog(r,np.abs(a_prime(r_normal)),'-',lw=2,label=r'$|a^\prime|$')

plt.loglog(r,np.abs(a_dot(r_normal)),'-',lw=2,label=r'$|\dot a|$')

plt.loglog(r,np.abs(a_prime(r_normal) - a_dot(r_normal)),':',lw=2,label=r'$|a^\prime - \dot a|$')


plt.xlabel('log (r [kpc])', fontsize=12)
plt.ylabel('log(Axion field Amplitude)', fontsize=12)

plt.title(f"Axion profile based on NFW and derivatives at $m_a={m_a:.0e}$ eV")

plt.grid(True, which='both', alpha=0.3)
plt.legend()
plt.tight_layout()

plt.show()
```
