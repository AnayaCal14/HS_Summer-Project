```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#nearby clusters- virgo, fornax, coma
params1 = {
    'rho_0': 1e4,
    'r_s': 340,
    'b': 3,}
#massive clusters- El gordo, perseus, bullet, Abell 8  
params2 = {
    'rho_0': 5e4,
    'r_s': 500,
    'b': 3.2}

# gNFW
def gnfw(r, rho_0, r_s, b):
    alpha = 1
    gamma = 1
    x = r / r_s
    return rho_0 / ((x**gamma) * (1 + x**alpha)**((b-gamma)/alpha))

# radius range
r = np.logspace(-1, 4, 5000)

# densities
rho1 = gnfw(r, **params1)
rho2 = gnfw(r, **params2)

# plot
plt.figure(figsize=(8,6))

plt.loglog(
    r, rho1,
    linewidth=2,
    label='Nearby Clusters'
)

plt.loglog(
    r, rho2,
    linewidth=2,
    label='Massive Clusters'
)

plt.xlabel(r'log[r (kpc)]', fontsize=13)
plt.ylabel(r'Density $\rho(r)$', fontsize=13)
plt.title('gNFW Density Profiles', fontsize=14)

plt.legend()

plt.grid(
    True,
    which='both',
    linestyle='--',
    alpha=0.4
)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# gNFW
def gnfw(r, rho_0, r_s, b):

    alpha = 1
    gamma = 1

    x = r / r_s

    return rho_0 / (
        (x**gamma) *
        (1 + x**alpha)**((b-gamma)/alpha)
    )

#radius r
r = np.logspace(-1, 3, 2000)

#parameter ranges

# Nearby clusters range
rho0_range1 = [1e4, 5e4]
rs_range1   = [300, 500]
b_range1    = [3, 3.3]

# Massive clusters range
rho0_range2 = [5e4, 2e5]
rs_range2   = [500, 800]
b_range2    = [3.2, 3.6]

# minmax

# Nearby clusters
rho1_min = gnfw(r,
                rho0_range1[0],
                rs_range1[0],
                b_range1[0])

rho1_max = gnfw(r,
                rho0_range1[1],
                rs_range1[1],
                b_range1[1])

# Massive clusters
rho2_min = gnfw(r,
                rho0_range2[0],
                rs_range2[0],
                b_range2[0])

rho2_max = gnfw(r,
                rho0_range2[1],
                rs_range2[1],
                b_range2[1])

# plot
plt.figure(figsize=(8,6))

# Nearby band
plt.fill_between(
    r,
    rho1_min,
    rho1_max,
    alpha=0.4,
    label='Nearby Clusters Range'
)

# Massive band
plt.fill_between(
    r,
    rho2_min,
    rho2_max,
    alpha=0.4,
    label='Massive Clusters Range'
)


plt.loglog()

plt.xlabel(r'log[r (kpc)]', fontsize=13)
plt.ylabel(r'Density $\rho(r)$', fontsize=13)

plt.title('gNFW Density Profile Ranges', fontsize=14)

plt.legend()

plt.grid(
    True,
    which='both',
    linestyle='--',
    alpha=0.4
)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# gNFW PROFILE

def gnfw(r, rho_0, r_s, b):

    alpha = 1
    gamma = 1

    x = r / r_s

    return rho_0 / (
        (x**gamma) *
        (1 + x**alpha)**((b-gamma)/alpha)
    )

#radius
r = np.logspace(-1, 3, 2000)

# range

# Nearby clusters
rho0_range1 = [1e4, 5e4]
rs_range1   = [300, 500]
b_range1    = [3.0, 3.3]

# Massive clusters
rho0_range2 = [5e4, 2e5]
rs_range2   = [500, 800]
b_range2    = [3.2, 3.6]

# minmax

rho1_min = gnfw(r, rho0_range1[0], rs_range1[0], b_range1[0])
rho1_max = gnfw(r, rho0_range1[1], rs_range1[1], b_range1[1])

rho2_min = gnfw(r, rho0_range2[0], rs_range2[0], b_range2[0])
rho2_max = gnfw(r, rho0_range2[1], rs_range2[1], b_range2[1])

# means
rho1_mean = 0.5 * (rho1_min + rho1_max)
rho2_mean = 0.5 * (rho2_min + rho2_max)

rho_mean = 0.5 * (rho1_mean + rho2_mean)


plt.figure(figsize=(9,6))

# Shaded regions
plt.fill_between(r, rho1_min, rho1_max, alpha=0.25, color='red', label='Nearby range')
plt.fill_between(r, rho2_min, rho2_max, alpha=0.25, color='green', label='Massive range')

# Mean lines
plt.loglog(r, rho1_mean, color='red', linewidth=1, linestyle='--', label='Nearby mean')
plt.loglog(r, rho2_mean, color='green', linewidth=1, linestyle='--', label='Massive mean')

# Combined mean
plt.loglog(r, rho_mean, color='black', linewidth=1, label='Global mean')

#
plt.xlabel(r'Radius $r$ (kpc)', fontsize=13)
plt.ylabel(r'Density $\rho(r)$', fontsize=13)
plt.title('gNFW Density Profile: Ranges + Mean', fontsize=13)

plt.grid(True, which='both', linestyle='--', alpha=0.4)
plt.legend()

plt.tight_layout()
plt.show()
```

DENSITY PROFILE FOR COMA CLUSTER : 


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#coma
#rho_0 = 1e4 rh0_c, rho_c = 10^-29 g/cm3 
#r_s = 340 kpc
#b = 1   #they took a very low value in the paper , but lets seee 

params1 = {
    'rho_0': 1e4,
    'r_s': 340, #kpc,
    }

# gNFW
def gnfw(r, rho_0, r_s):
    
    x = r / r_s
    return rho_0 / ((x) * (1 + x)**(2))

# radius range
r = np.logspace(-1, 3, 5000)

# densities
rho1 = gnfw(r, **params1)


# plot
plt.figure(figsize=(8,6))

plt.loglog(
    r, rho1,
    linewidth=2,
    label='Coma Cluster'
)


plt.xlabel(r'log[r (kpc)]', fontsize=13)
plt.ylabel(r'Density $\rho(r)$', fontsize=13)
plt.title('gNFW Density Profile', fontsize=14)

plt.legend()

plt.grid(
    True,
    which='both',
    linestyle='--',
    alpha=0.4
)

plt.tight_layout()
plt.show()
```

```python
#coma
#rho_0 = 1e4 rh0_c
#r_s = 340 kpc
#b = 1

#NFW profile

def gnfw(r, rho_0, r_s):

    x = np.abs(r) / r_s   # use absolute radius

    return rho_0 / (x * (1 + x)**2)

r_left  = np.logspace(-3, 3, 2500) * (-1)
r_right = np.logspace(-3, 3, 2500)

# combine
r = np.concatenate((r_left[::-1], r_right))

rho1 = gnfw(r, **params1)

#P;-

plt.figure(figsize=(9,6))

plt.plot(
    r,
    rho1,
    linewidth=2,
    label='Coma Cluster'
)

# logarithmic y-axis only
plt.yscale('log')

plt.xlabel(r'Radius $r$ (kpc)', fontsize=13)
plt.ylabel(r'Density $\rho(r)$', fontsize=13)

plt.title('Two-Sided gNFW Density Profile', fontsize=14)

plt.axvline(0, color='black', linestyle='--', alpha=0.6)

plt.legend()

plt.grid(
    True,
    which='both',
    linestyle='--',
    alpha=0.4
)

plt.tight_layout()
plt.show()
```

```python
#coma
#rho_0 = 1e4 rh0_c
#r_s = 340 kpc
#b = 1

#NFW profile

def gnfw(r, rho_0, r_s):

    x = np.abs(r) / r_s   # use absolute radius

    return rho_0 / (x * (1 + x)**2)

Rmax = 1e3 #kpc
r = np.linspace(0, 2*Rmax, 5000)

rho1 = gnfw(r, **params1)

#Plot

plt.figure(figsize=(9,6))

plt.loglog(
    r,
    rho1,
    linewidth=2,
    label='Coma Cluster'
)


plt.xlabel(r'Radius $r$ (kpc)', fontsize=13)
plt.ylabel(r'Density $\rho(r)$', fontsize=13)

plt.title('Two-Sided gNFW Density Profile', fontsize=14)

plt.axvline(0, color='black', linestyle='--', alpha=0.6)

plt.legend()

plt.grid(
    True,
    which='both',
    linestyle='--',
    alpha=0.4
)

plt.tight_layout()
plt.show()
```

```python
def gnfw_shifted(r, rho_0, r_s, center):

    x = np.abs(r - center) / r_s

    return rho_0 / (x * (1 + x)**2)

# shifted radius 

Rmax = 1e3   # kpc

r = np.linspace(0, 2*Rmax, 5000)

# cluster center at Rmax
rho1 = gnfw_shifted(r, center=Rmax, **params1)

#Plot

plt.figure(figsize=(9,6))

plt.plot(
    r,
    rho1,
    linewidth=2,
    label='Coma Cluster'
)

plt.yscale('log')

plt.xlabel(r'Radius $r$ (kpc)', fontsize=13)
plt.ylabel(r'Density $\rho(r)$', fontsize=13)

plt.title('Shifted Two-Sided NFW Density Profile', fontsize=14)

# mark center
plt.axvline(Rmax, color='black', linestyle='--',
            alpha=0.7, label='Cluster Center')

plt.legend()

plt.grid(
    True,
    which='both',
    linestyle='--',
    alpha=0.4
)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#rho_0 = 1.74 · 10−2 Msun / pc^3
#r_s = 12.5 kpc
#b = 1

#DM parameters 
rho_0_SI = 1.74 * 1e-2 * 6.77 * 1e-20


params1 = {
    'rho_0': rho_0_SI ,#Msun/pc^3
    'r_s': 12.5, #kpc,
    }


# radius range
Rmax = 146.5 #kpc
r = np.linspace(0, 2*Rmax, 5000)
centre = Rmax
# gNFW
def gnfw(r, rho_0, r_s):
    
    x = np.abs(r - centre) / r_s
    return rho_0 / ((x) * (1 + x)**(2))


# densities
rho1 = gnfw(r, **params1)


# plot
plt.figure(figsize=(8,6))

plt.loglog(
    r, rho1,
    linewidth=2,
    label='Andromeda'
)


plt.xlabel(r'log[r (kpc)]', fontsize=13)
plt.ylabel(r'Density $\rho(r)$', fontsize=13)
plt.title('gNFW Density Profile', fontsize=14)

plt.legend()

plt.grid(
    True,
    which='both',
    linestyle='--',
    alpha=0.4
)

plt.tight_layout()
plt.show()
```
